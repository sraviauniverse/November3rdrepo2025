Now, let's understand how can we take input from the user.

In python , input() function is used to prompt the user for input during the execution of a program.
It allows you to interact with the user by receiving input from them and using that input within your program.

When input() is called, it displays a message or prompt to the user, and the program waits for the user to enter a value followed by pressing the Enter key.

Once the user enters the input, it is captured as a string and can be assigned to a variable or used directly in your code.

```python
First_name= input()
```

```python
print(First_name)
```

Lets check the type of the input text

```python
type(First_name)
```

Lets define a function which adds two numbers.

```python
def add_numbers(a, b):
    return a + b
```

In the above cell, we come across 'return'. The return statement in Python is used to specify the value that a function should return after executing its logic.

It allows you to pass back a result or output from the function to the caller.

```python
result = add_numbers(5, 3)
```
