### LIST
  * Lists are used to store multiple items in a single variable

  * List items are ordered, changeable, and allow duplicate values.

  * List items are indexed, the first item has index [0], the second item has    index [1] etc.

```python
# list
l = ['apple',1,3,'orange',4.0,True ]
l
```

```python
dir(l)
```

```python
# Adding element at the last of a list
l.append('mango')
```

```python
#The clear() method empties the list.
#The list still remains, but it has no content.
#l.clear()
#l
```

You cannot copy a list simply by typing list2 = list1, because: list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.

There are ways to make a copy, one way is to use the built-in List method copy().

```python
# copy
l1 = l.copy()
l1
```

count() is used to count the element in the list

```python
l1.count(3)
```

To append elements from another list to the current list, use the extend() method.

```python
l.extend(l1)
l
```

List items are indexed and you can access them by referring to the index number:

```python
l[2:4]
```

```python
# this returs the index of the element
l.index(3)
```

To insert a list item at a specified index, use the insert() method.

The insert() method inserts an item at the specified index:

```python
# insert is used for inserting the element in the list
l.insert(1,25)#listname.insert(index, element)
print(l)
```

```python
# pop() method is used for removing the last element in the list
l2=l.pop()
l2
```

```python
# remove() method is used for removing the specific element in the list and only first occurence of the elemetn is removed
l1.remove('apple')
l1
```

```python
# reversing the element in the list
l1.reverse()
l1
```

```python
# used for sorting
l1 = [3,4,2,4,6,2,1]
l1.sort()
```
