Now let us check out format() function.

The format() method in Python is used to format a string by replacing placeholders {} with corresponding values.

It supports various formatting options to control alignment, precision, number representation, and more, providing flexibility in generating formatted output

Let's work with some examples.

```python
value = "Hello"
```

If you want to left align it:

```python
formatted = "{:<10}".format(value)
print(formatted)
```

If you want to right align the result, then

```python
formatted = "{:>10}".format(value)
print(formatted)
```

If you want to center align the result, then

```python
formatted = "{:^10}".format(value)
print(formatted)
```

If you want to format a number to binary, then

```python
value2 = 10
```

```python
formatted = "{:b}".format(value2)
print(formatted)
```

If you want format a number to its corresponding unicode character, then

```python
value3 = 65
```

```python
formatted = "{:c}".format(value3)
print(formatted)
```

If you want to format a value to octal, then

```python
formatted = "{:o}".format(value3)
print(formatted)
```

If you want to use a plus sign for positive or negative values, then

```python
formatted = "{:+}".format(value3)
print(formatted)
```
