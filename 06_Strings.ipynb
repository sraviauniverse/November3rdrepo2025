What if I want to Reverse the string?

In python, we don't have any in-built function to reverse a string, for reversing, we will use the same slicing operator.

Here's how we do it.

```python
info[-1: :-1]# start index is -1, Grab everything, but go in steps size of -1
```

```python
len(info)
```

We can do the reversing of a sting in 3 different way.

Here's how we can do it:

```python
info[18: : -1]# start index is 18, Grab everything, but go in steps size of -1
```

```python
info[len(info): :-1]# start index is length of string, Grab everything, but go in steps size of -1
```

```python
info[: :-1]#We can use this to print a string backwards
```

Now, lets try replacing substrings in a string using replace().

This method replaces a segment of the string, i.e. a substring with a new string. We input the part of the string we would like to change.

The second argument is what we would like to exchange the segment with, and the result is a new string with the segment changed:

```python
A="Michael Jackson is the best"
B=A.replace('Michael', 'Janet')
B
```

What if we want to find whether a substring is present in the string?

To find that, we use find() function.

This method "find" finds a sub-string. The argument is the substring you would like to find, and the output is the first index of the sequence.

We can find the sub-string "jack" or "el".

```python
Name="Michael Jackson"
Name.find('el')
```

Similarly, if we want to find whether a string starts with certain substring or ends with a certain substring?

For that, we use startswith() and endswith() function.

Here is the example:

```python
Name.startswith('Mi')
```

If startswith(), condition has met, then it will return boolean value "True", otherwise "False"

```python
Name.startswith('MI')
```

From the above about, we can say that, to find whether a string startswith or endswith specific characters, we should be very careful with the cases.

```python
Name.endswith('son')
```
